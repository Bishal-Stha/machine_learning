```python
import pandas as pd
import numpy as np

df_train = pd.read_csv('./titanic/train.csv')
df_test = pd.read_csv('./titanic/test.csv')
df_val = pd.read_csv('./titanic/gender_submission.csv')
df_test['Survived'] = df_val['Survived']
```

```python
print(df_train.shape)
print(df_test.shape)
```

```python
df_train['Sex'] = df_train['Sex'].map({'female':0.0,'male':1.0})
df_test['Sex'] = df_test['Sex'].map({'female':0.0,'male':1.0})
df_train['Embarked'] = df_train['Embarked'].map({'S':0, 'Q':1, 'C':2})
df_test['Embarked'] = df_test['Embarked'].map({'S':0, 'Q':1, 'C':2})
```

```python
df_test.drop(columns=['Name','Ticket','Cabin'],inplace=True)
df_train.drop(columns=['Name','Ticket','Cabin'], inplace=True)
df_train.dropna(inplace=True)
df_test.dropna(inplace=True)
df_train.drop_duplicates(inplace=True)
df_test.drop_duplicates(inplace=True)
```

```python
X_train = df_train.drop(columns=['Survived'])
y_train = df_train['Survived']
X_test = df_test.drop(columns=['Survived'])
y_test = df_test['Survived']
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

```python
X_test = df_test[X_train.columns]
```

```python
print(X_train.columns)
print(X_test.columns)
```

```python
from sklearn.linear_model import LogisticRegression
# Train model
logreg = LogisticRegression()
logreg.fit(X_train, y_train)

# Predict
y_pred = logreg.predict(X_test)
print(y_pred)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
